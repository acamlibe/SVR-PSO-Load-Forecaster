```python
import numpy as np
import pandas as pd
from sklearn.svm import SVR
```

```python
load_df = pd.read_csv('Data/DOM/Load Actuals/Processed/Aggregated/load.csv')
load_df
```

```python
calendar_df = pd.read_csv('Data/calendar.csv')
calendar_df
```

```python
df = pd.concat([load_df, calendar_df]).groupby('DATE').mean()

df.info()
```

```python
df_usable = df[168:]
df_usable
```

```python
for i in range(1, 169):
    df_usable['lag#' + str(i)] = df['MW'].shift(i, fill_value=0)

df_usable
```

```python
x = df.iloc[:, 1:]
y = df.iloc[:, 0]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
model = SVR(kernel='rbf', C=1000.0, gamma=0.15)
model.fit(x_train, y_train)
```

```python
pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_percentage_error

mean_absolute_percentage_error(y_test, pred)
```

```python
from sklearn.model_selection import GridSearchCV

svm_params = {
  'C': [0.01, 0.1, 1, 10, 100],
  'gamma': [0.001, 0.01, 0.1, 1],
  'kernel': ['rbf']
}
svm_grid = GridSearchCV(model, param_grid=svm_params,
 refit=True, verbose=0, cv=2)

svm_grid.fit(x_train, y_train)
svm_grid.best_params_
```

```python

from sklearn.metrics import mean_absolute_percentage_error

model_optimized = SVR(kernel='rbf', C=100.0, gamma=1)
model_optimized.fit(x_train, y_train)
pred_optimized = model_optimized.predict(x_test)
mean_absolute_percentage_error(y_test, pred_optimized)
```
